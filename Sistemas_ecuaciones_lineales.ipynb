```python
import numpy as np

# Función para resolver un sistema de ecuaciones 2x2
def resolver_sistema_2x2(a, b, c, d, e, f):
    # Matriz de coeficientes
    A = np.array([[a, b], [d, e]])
    # Matriz de términos independientes
    B = np.array([c, f])

    # Intentamos resolver el sistema
    try:
        solucion = np.linalg.solve(A, B)
        x, y = solucion
        return f"Solución: x = {x}, y = {y}"
    except np.linalg.LinAlgError:
        return "El sistema no tiene solución única (puede ser dependiente o inconsistente)."

# Ejemplo de uso
a=int(input("Ingrese el valor de a: "))
b=int(input("Ingrese el valor de b: "))
c=int(input("Ingrese el valor de c: "))
d=int(input("Ingrese el valor de d: "))
e=int(input("Ingrese el valor de e: "))
f=int(input("Ingrese el valor de f: "))

'''a, b, c = 3, 2, 1
d, e, f = 1, 4, 6'''

print(resolver_sistema_2x2(a, b, c, d, e, f))
```

```python
import numpy as np

# Función para resolver un sistema de ecuaciones 3x3
def resolver_sistema_3x3(a, b, c, d, e, f, g, h, i, j, k, l):
    # Matriz de coeficientes
    A = np.array([[a, b, c],
                  [d, e, f],
                  [g, h, i]])

    # Matriz de términos independientes
    B = np.array([j, k, l])

    # Intentamos resolver el sistema
    try:
        solucion = np.linalg.solve(A, B)
        x, y, z = solucion
        return f"Solución: x = {x}, y = {y}, z = {z}"
    except np.linalg.LinAlgError:
        return "El sistema no tiene solución única (puede ser dependiente o inconsistente)."

# Ejemplo de uso
a, b, c, d, e, f, g, h, i, j, k, l = 2, -1, 3, 1, 3, 2, 1, -1, -2, 5, 12, -3
print(resolver_sistema_3x3(a, b, c, d, e, f, g, h, i, j, k, l))
```
